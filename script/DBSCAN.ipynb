```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('clean_iqr.csv')

df
```

```python
training_set = df[df.columns[1:]]

training_set
```

```python
from sklearn.cluster import DBSCAN


X = training_set.to_numpy()

X
```

```python
from sklearn.metrics import silhouette_score as ss
```

```python
import itertools

epsilons = np.linspace(0.03, 1.3, num=30)
min_samples = np.arange(2, 20, step=1)

combinations = list(itertools.product(epsilons, min_samples))

combinations
```

```python
from sklearn.preprocessing import StandardScaler

X_scaled = StandardScaler().fit_transform(X)

X_scaled
```

```python
X_scaled.shape
```

```python
def get_scores_and_labels(combinations, X):
  scores = []
  all_labels_list = []
  all_num_cluster_list = []

  for i, (eps, num_samples) in enumerate(combinations):
    dbscan_cluster_model = DBSCAN(eps=eps, min_samples=num_samples).fit(X)
    labels = dbscan_cluster_model.labels_
    labels_set = set(labels)
    num_clusters = len(labels_set)
    all_num_cluster_list.append(num_clusters)
    if -1 in labels_set:
      num_clusters -= 1

    if (num_clusters < 2) or (num_clusters > 10):
      scores.append(-10)
      all_labels_list.append('bad')
      c = (eps, num_samples)
      print(f"Bad -- {c}: num_cluster: {num_clusters}")
      continue

    scores.append(ss(X, labels))
    all_labels_list.append(labels)
    print(f"Score: {scores[-1]}, Labels: {all_labels_list[-1]}, NumClusters: {num_clusters}")

  best_index = np.argmax(scores)
  best_parameters = combinations[best_index]
  best_labels = all_labels_list[best_index]
  best_score = scores[best_index]
  best_num_cluster = all_num_cluster_list[best_index]

  return {'best_epsilon': best_parameters[0],
          'best_min_samples': best_parameters[1],
          'best_labels': best_labels,
          'best_score': best_score,
          'num_clusters': best_num_cluster}
```

```python
best_dict = get_scores_and_labels(combinations, X_scaled)
```

```python
best_dict
```

```python
df['cluster'] = best_dict['best_labels']

df
```
